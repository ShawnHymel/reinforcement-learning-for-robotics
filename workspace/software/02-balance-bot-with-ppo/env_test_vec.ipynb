```python
# Import standard libraries
from pathlib import Path
import sys
import time

# Third-party libraries
import gymnasium as gym
from gymnasium.utils.env_checker import check_env
from gymnasium.wrappers import RecordEpisodeStatistics
import numpy as np
```

```python
# Add the folder containing our envs/ and rl/ packages to the path
sys.path.append("/workspace/software")

# Import the custom environment
from envs.balance_bot_env import BalanceBotEnv
```

```python
# Settings
MJCF_PATH = Path("/workspace/mechanical/FreeCAD/bala2-fire/bala2-fire-simplified.xml")
TIMESTEP = 0.002
SEED = 42
```

```python
# Try wrapping the environment in a vector
wrapped_env = RecordEpisodeStatistics(
    BalanceBotEnv(mjcf_path=MJCF_PATH, render_mode="human")
)
vec_envs = gym.vector.SyncVectorEnv([lambda: wrapped_env])
```

```python
# Test: random policy
obs, _ = vec_envs.reset()
total_reward = 0

# Try a random policy: move the motors in random directions/speeds
terminated = False
terminated = False
for step in range(3500):
    step_start = time.time()

    # Choose a random action
    action = vec_envs.action_space.sample()

    # Take a step with the given action
    obs, reward, terminated, truncated, _ = vec_envs.step(action)

    # Accumulate reward
    total_reward += reward

    # Render in MuJoCo
    vec_envs.envs[0].render()

    # Log steps
    print(f"step {step} OK")

    # Show that we stopped
    if terminated or truncated:
        print(f"Episode ended at step {step}")
        break

    # Make sure the loop iteration time matches the MJCF timestep time (2 ms)
    # i.e. slow the simulation down so we can see the robot behave in real time
    slack = TIMESTEP - (time.time() - step_start)
    if slack > 0:
        time.sleep(slack)

print(f"Final obs: {obs}")
print(f"Final reward: {total_reward}")
```

```python
# Close the environment
vec_envs.close()
```
