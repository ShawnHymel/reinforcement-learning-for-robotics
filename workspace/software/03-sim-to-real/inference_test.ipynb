```python
from pathlib import Path

import onnxruntime as ort
import numpy as np
```

```python
# Settings
ONNX_PATH = Path("../02-balance-bot-with-ppo/runs/BalanceBot-v0__balance-bot-ppo__42__1778105482/actor.onnx")

# Test observation [pitch, pitch_rate, wheel_vel_left, wheel_vel_right]
TEST_OBS = np.array([[0.2, 1.0, -0.5, 0.5]], dtype=np.float32)
```

```python
# Load ONNX model
session = ort.InferenceSession(ONNX_PATH)

# Run inference (input and output names must match names on model)
result = session.run(["action"], {"observation": TEST_OBS})

# Print result
print(result)
```
